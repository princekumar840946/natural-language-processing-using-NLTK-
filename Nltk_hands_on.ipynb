Now we use frequency distribution

```
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg
```

```
sample=gutenberg.raw("carroll-alice.txt")
```

```
word_tokens=word_tokenize(sample)
```

```
freq
```

```
freq.most_common(20)
```

```
freq.plot(20)
```

#stopwords  are just useless words, which has no signficant information 

```
from nltk.corpus import stopwords

stop_words=set(stopwords.words("english"))
print(stop_words)
```

to remove english stopwords from our sample

```
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import genesis
sample=genesis.raw("english-web.txt")
word_tokens=word_tokenize(sample)
word_tokens
```

```
filtered_words=[w for w  in word_tokens if w not in stop_words and len(w)>3]
```

```
filtered_words
```

```
freq=FreqDist(filtered_words)
freq.most_common(20)
```

Thus the above result gives significant information about the text

```
freq.plot(30)
```

Bigrams:- these are two consecutive words that appear in text

```
import nltk
nltk.download('webtext')
  
from nltk.corpus import webtext
from nltk import bigrams

```

we have to make it lower caps otherwise if one word is upper caps and one is lower it will be treated two diff words

```
words=[w.lower() for w in webtext.words("pirates.txt") if w not in stop_words and len(w)>3]
```

```
bigram=bigrams(words)
```

```
freq=FreqDist(bigram)
```

```
freq.most_common(20)
```

```
freq.plot(20)
```

Trigrams:-it gives  us three consecutive words in a text

```
from nltk import trigrams
words=[w.lower() for w in webtext.words("grail.txt")if w not in stop_words and len(w)>3]
```

```
trigram=trigrams(words)
```

```
freq=FreqDist(trigram)
```

```
freq.most_common(20)
```

```
freq.plot(20)
```

Ngrams:-these are N continous words in a text

```
from nltk import ngrams
from nltk.corpus import gutenberg 
```

```
words = [w.lower() for w in gutenberg.words("carroll-alice.txt") if w not in stop_words and len(w)>3 ]
```

```
ngram =ngrams(words,4)
```

```
freq=FreqDist(ngram)
```

```
freq.most_common(20)
```

```
freq.plot(20)
```

Stemming :-its a process for text  noemalization which involves reducing words to their root base .


```
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
nltk.download('state_union')
  
from nltk.corpus import state_union
```

```
words = word_tokenize(state_union.raw("1961-Kennedy.txt"))
```

```
ps=PorterStemmer()
```

```
for w in words:
  print(ps.stem(w))
```

Lemmanization:- its a process which involves reducing words to their root using vocabulary.
Its meaningful which sometimes not happen in stemming

```
from nltk.stem import WordNetLemmatizer
```

```
ps=WordNetLemmatizer()
```

```
words=["cats","feets","bats","president","whoose","prime minister"]
for w in words:
  print(ps.lemmatize(w))
```
